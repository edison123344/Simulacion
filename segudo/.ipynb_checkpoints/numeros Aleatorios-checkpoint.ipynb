```python


import collections
def division(long,numeroOriginal,digitos):
    numero=int((long/2)-(digitos/2)+1)
    sNumero=str(numeroOriginal)
    numUI=sNumero[numero-1:numero+3]
    return int(numUI)


def buscarSemilla(valor,lista):
    for x in range(0,len(lista)):
        if (lista[x]==valor):
               return True

def generarAleatorios(iteraciones,semilla):
    iteraciones=iteraciones
    Xo=semilla
    digitos=4
    Xn=0
    XnPo=0
    longuitud=0
    Ui=0
    Rn=0
    arregloSemillas=[]
    semillasRepetidas=[]
    
    if len(str(Xo))>3:
        print('Resultado')
        for i in range(iteraciones):
            if (i==0):
                Xn=Xo
                XnPo=Xn**2
                longuitud=len(str(XnPo))
                Ui=division(longuitud,XnPo,digitos)
                arregloSemillas.append(Xn)
                Rn=Ui/10000
                print(Rn) 
            else:
                Xn=Ui
                XnPo=Xn**2
                longuitud=len(str(XnPo))
                Ui=division(longuitud,XnPo,digitos)
                Rn=Ui/10000
                print(Rn)
                busquedaSemilla=buscarSemilla(Xn,arregloSemillas)
                if(busquedaSemilla==True):
                     break           
                arregloSemillas.append(Xn)

generarAleatorios(100,68977)
```

### NUMEROS PSEUDOALEATORIOS  OBTENIDOS DE LAS ESTADISTICAS DEL CPU y RAM



```python
import psutil as coretemp
import cpuinfo as cpu

class NumPseudoaleatorios:
    def __init__(self, iterator, cifras):
        self.iterator = iterator
        self.cifras = cifras
    # Algoritmo de cuadrados medios
    def pos(self,cifras):
        valor1 =0
        valor2 =0
        if cifras%2 !=0:
            valor1 = int(cifras/2)
            valor2 = int(cifras/2)+1
        valor1 = int(cifras/2)
        valor2 = int(cifras/2)
        return valor1,valor2

    def get_numero(self,iterator, val, cifras):
        x0_semilla = int(val)
        aum = self.pos(cifras)
        print("N-Iter","  Xn  ","  Xn*Xn   ", " Longitud "," Ui"," Rn ")
        for i in range(iterator):
            xn2= x0_semilla**2
            long = len(str(xn2))
            ui = str(xn2)[int(long/2)-aum[0]:int(long/2)+aum[1]]
            rn = int(ui)/10**cifras
            print(i,"        "," ",x0_semilla," ",xn2, " ", long, " ", ui, " ", rn)
            x0_semilla=int(ui)
        print("   ")
    
    def execute(self,param):
        for i in param:
            print("Semilla: ", i)
            self.get_numero(self.iterator, i, self.cifras)
            print("  ")

if __name__ == "__main__":
 
    cpu_uso=coretemp.cpu_percent()
    ram=coretemp.virtual_memory().total
    ram_uso=coretemp.virtual_memory().used
    params = [ cpu_uso, ram, ram_uso]
    
    nums = NumPseudoaleatorios(20,2)
    nums.execute(params)
```
